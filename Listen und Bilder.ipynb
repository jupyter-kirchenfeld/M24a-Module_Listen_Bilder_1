## NumPy Array

Das wichtigste Objekt, welches das NumPy Modul zur Verfügung stellt ist das n-dimensionale Array `ndarray`. Wir haben bereits mit Arrays gearbeitet ohne dies zu Wissen! Den eine Liste ist nichts anders als ein eindimensionales Array. Weiter unten werden wir uns mit zweidimensionalen Arrays beschäftigen doch zu erst wollen wir NumPy importieren und unseren vorherrigen Code zum Erstellen von Listen mit zufälligen Elementen mittels NumPy realiseren.

Als erstes müssen wir wieder NumPy importieren:

```python
import numpy as np
```

das Schlüsselwort `as` macht uns das Modul `numpy` unter dem Namen `np` verfügbar. Dies ist eine Konvention, es ist nicht notwendig das Modul umzubenennen, aber bei den meisten Tutorials und Anleitungen aus dem Internet hat sich diese Namensgebung durchgesetzt.

Danach können wir mittels der Funktion [randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html?highlight=randint#numpy-random-randint) aus dem Objekt `random` analog zur Funktion [randint](https://docs.python.org/3/library/random.html#random.randint) aus dem Built-in Modul `random` Zufallszahlen generieren:

```python
np.random.randint(10)
```

Im Unterschied zu der Funktion aus dem Built-in Modul können wir aber auch direkt Listen erstellen:

```python
np.random.randint(10, size=5)
```

mit dem Parameter `size` können wir der Funktion mitteilen, wie gross die Liste sein soll.

## Array als Bild interpretieren

Nun wollen wir unser Wissen über [Rastergrafiken](https://informatik.mygymer.ch/base/?page=code/4-graphics/1-basics/) mit dem NumPy Modul verknüpfen um aus zweidimensionalen Listen Bilder zu generieren. Dazu erstellen wir mittels `np.random.randint` eine zweidimensionale Liste:

```python
listen_in_liste = np.random.randint(10, size=(5, 5))
listen_in_liste
```

**Bemerkungen:**

- Erinnern Sie sich an die verschachtelten Listen? 
- Die Syntax `(5,5)` beschreibt einen sogenannte [Tupel](https://docs.python.org/3/library/stdtypes.html#tuple) wir werden hier nicht weiter darauf eingehen

Mittels des Modul [matplotlib](https://matplotlib.org/stable/api/index) können wir uns nun diese zweidimensionale Liste mit Zufallszahlen als Bild darstellen:

```python
%matplotlib inline
import matplotlib
matplotlib.pyplot.imshow(listen_in_liste, cmap = 'gray')
```
